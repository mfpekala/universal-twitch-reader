```python
import ffmpeg
```

```python
def convert2frames(video_path, output_path, fps=1/2): # Default to one frame every two seconds
    probe = ffmpeg.probe(video_path)
    print(probe['streams'][0].keys())
    time = float(probe['streams'][0]['duration'])
    num_frames = int(time * fps)
    interval = 1 / fps

    offsets = [i * interval for i in range(num_frames)]
    
    for (ix, offset) in enumerate(offsets):
        (
            ffmpeg
                .input(video_path, ss=offset)
                .filter('scale', 1980, -1)
                .output(f"{output_path}/f_{int(ix * interval * 1000)}.jpg", vframes=1)
                .run()
        )
        print(ix)
    

convert2frames("videos/league/jankos.mp4", "images/league/jankos")
```
